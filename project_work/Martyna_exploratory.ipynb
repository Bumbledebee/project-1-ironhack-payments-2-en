```python
#creating a column with YYY-MM for activity month
df_cash["activity_month"] = df_cash["created_at"].str[:7] + "-01"
```

```python
df_cash.head()
```

```python

#changing dtype to datetime64
df_cash["activity_month"] = df_cash["activity_month"].astype("datetime64[s]")
```

```python
#cohort_month -> the min from activity_month

df_cash["cohort_month"] = df_cash["user_id"].map(df_cash.groupby("user_id")["activity_month"].min()) 
```

```python
#adding new column with cohort index
df_cash["cohort_index"] = (df_cash["activity_month"].dt.to_period("M").astype(int) - df_cash["cohort_month"].dt.to_period("M").astype(int))
```
