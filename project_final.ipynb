### Creating the Columns for the Cohorts

```python
# creating a column with YYYY-MM-DD for the activity month. 
# activity month is the month is which the cash request was made

cash["activity_month"] = cash["created_at"].str[:7] + "-01"
```

```python
cash[["id", "created_at", "activity_month"]]
```

```python
#checking the dtype of activity month
cash[["id", "created_at", "activity_month"]].dtypes
```

```python
# making sure that we have a datetime object for activity month
cash["activity_month"] = cash["activity_month"].astype("datetime64[s]")
```

```python
#checking activity month has changed dtype to datetime
cash[["id", "created_at", "activity_month"]].dtypes
```

```python
# creating cohorts
# users are assigned to the cohort of their first activity.
cash["cohort_month"] = cash["user_id"].map(cash.groupby("user_id")["activity_month"].min()) 
```

```python
# adding a new column with cohort as index
# looking to group the activites to the month after the cohort started
cash["cohort_index"] = (cash["activity_month"].dt.to_period("M").astype(int) \
                           - cash["cohort_month"].dt.to_period("M").astype(int))
```

```python
# looking at the columns needed
cash[["id", "created_at", "user_id", "activity_month", "cohort_month", "cohort_index"]]
```

```python
#find a user with the maximum frequency of cash requests
cash["user_id"].value_counts()
```

```python
# creating a dataframe for the user with the maximum cash requests
user_max = cash[cash["user_id"] == 3377][["id", "created_at", "user_id", "activity_month", "cohort_month", "cohort_index"]]
```

```python
# looking at the data for that user sorted in descending order
user_max.sort_values(by='activity_month', ascending = False)
```

As we can see the longer the activity period is the larger the cohort index will be.

### Merging Fees and Cash Requests
The id from the cash request file matches with cash_request_id from the fees file.

```python
#checking how many cash_request_ids from fees match with id from cash
matching_ids = cash[cash['id'].isin(fees['cash_request_id'])].shape[0]
#checking how many cash_request_ids do not match with id from cash
missing_ids_in_fees = cash[~cash['id'].isin(fees['cash_request_id'])].shape[0]
print(f"There are {matching_ids} matching cash request ids in fees. In comparison, there is a total \
number of {cash["id"].shape[0]} cash request ids. This makes a difference of {missing_ids_in_fees}")
```

The reason for more ids in cash compared to cash_request_id in fees might lie in the fact that a lot of cash requests got rejected or have not been paid back yet (no fees yet)

### Revenue Generated by the Cohort: 
Calculate the total revenue generated by each cohort over months to assess the financial impact of user behavior.

```python
#total fees by cash_request_id for all accepted payments
fees_agg_revenue = fees[fees["status"] == 'accepted'].groupby("cash_request_id")[["total_amount"]].sum()
fees_agg_revenue.reset_index()
```

```python
merged = cash.merge(fees_agg_revenue, how ='left', left_on='id', right_on='cash_request_id')
merged
```
